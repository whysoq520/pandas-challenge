## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
agg_dict_spender ={'Purchase ID': np.size,
          'Price': np.sum}

columns={'Purchase ID': 'Purchase Count',
         'Price': 'Total Purchase Value'}

top_spender = purchase_data.groupby('SN').agg(agg_dict_spender).rename(columns=columns)
top_spender
ave_purchase_price = top_spender["Total Purchase Value"]/top_spender["Purchase Count"]
top_spender["Average Purchase Price"] = ave_purchase_price
top_spender['Total Purchase Value'].dtype

top_spender = top_spender.sort_values('Total Purchase Value',ascending = False)

#op_spender["Average Purchase Price"]= top_spender["Average Purchase Price"].map("${:.2f}".format)
#tp_spender.rename(columns={'Price': 'Total Purchase Value'}) 

top_spender["Average Purchase Price"]= top_spender["Average Purchase Price"].map("${:.2f}".format)
top_spender[" Total Purchase Value"]= top_spender["Total Purchase Value"].map("${:.2f}".format)
top_spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
agg_dict_popitem ={'Purchase ID': np.size,
          'Price': np.sum}
        
columns={'Purchase ID': 'Purchase Count',
         'Price': 'Total Purchase Value'}

pop_item = purchase_data.groupby(['Item ID', 'Item Name']).agg(agg_dict_popitem).rename(columns=columns)
pop_item["Item Price"] = pop_item['Total Purchase Value']/pop_item['Purchase Count']
pop_item_2 = pop_item
pop_item_2 = pop_item_2.sort_values('Purchase Count', ascending = False).head(5)

pop_item_2["Total Purchase Value"]= pop_item_2["Total Purchase Value"].map("${:.2f}".format)
pop_item_2["Item Price"]= pop_item_2["Item Price"].map("${:.2f}".format)
pop_item_2.head()
```
